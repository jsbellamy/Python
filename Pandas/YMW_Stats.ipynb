```python
def sumthis(x):
    x1 = x.shift(-1)
    names = ['% Delta VS. Sector Total', '$ Delta VS. Sector Total', '% Delta VS. SPX Total', '$ Delta VS. SPX Total',
            'Active % Return', 'Sector % Return', 'SPX % Return', 'Alpha VS. Sector', 'Alpha VS. SPX']
    a = []
    a.append((x.iloc[:,0] - x.iloc[:,1])/x.iloc[:,1])
    a.append(x.iloc[:,0] - x.iloc[:,1])
    a.append((x.iloc[:,0] - x.iloc[:,2])/x.iloc[:,2])
    a.append(x.iloc[:,0] - x.iloc[:,2])
    a.append((x1.iloc[:,0] - x.iloc[:,0])/x.iloc[:,0])
    a.append((x1.iloc[:,1] - x.iloc[:,1])/x.iloc[:,1])
    a.append((x1.iloc[:,2] - x.iloc[:,2])/x.iloc[:,2])
    a.append(((x1.iloc[:,0] - x.iloc[:,0])/x.iloc[:,0])-(x1.iloc[:,1] - x.iloc[:,1])/x.iloc[:,1])
    a.append(((x1.iloc[:,0] - x.iloc[:,0])/x.iloc[:,0])-(x1.iloc[:,2] - x.iloc[:,2])/x.iloc[:,2])
    temp = pd.concat(a,axis=1)
    temp.columns = names
    temp.iloc[:,4:9] = temp.iloc[:,4:9].shift(1).fillna(0)
    results = pd.merge(x,temp, on = ['Date'])
    return results
```

```python
import pandas as pd
from functools import reduce
fileName = 'test1.xlsx'
df = []
for n in range(0,11):
    df.append(pd.read_excel(fileName,n,usecols = [0,11,12,13]).set_index('Date'))
df_final = reduce(lambda left,right: left.add(right,fill_value=0), df)
```

```python
year_range = []
i = 3
i1 = 3
for rows in ws.iter_rows(min_row=i, max_col=1):
    x = ws.cell(row=i, column=1).value.year
    x1 = ws.cell(row=i1, column=1).value.year
    i += 1
    if x != x1 or i == i1 + 1:
        year_range.append(ws.cell(row = i - 1, column = 1).value)
        i1 = i - 1

if ws.cell(row=i-1, column=1).value != year_range[-1]:
    year_range.append(ws.cell(row = i - 1, column = 1).value)
year_range[0:5]
```

```python
df_year = pd.DataFrame(year_range, columns = ['Date'])
df_year.tail()
```

```python
year_results = (pd.merge(df_year,df_final, on = 'Date').set_index('Date'))/10
year_results
```

```python
sumthis(year_results)
```
